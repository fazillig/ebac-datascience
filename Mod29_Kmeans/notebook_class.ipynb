## Carregando os dados
Vamos usar como exemplo a base de pinguins do seaborn. Como muitos algoritmos, o k-means no suporta valore faltantes (missings), ento vamos fazer uma substituio pelo valor mdio de cada varivel quantitativa. O k-means trabalha somente com variveis quantitativas.  possvel utilizar variveis dummy, mas h formas melhores de se trabalhar com variveis qualitativas.


````python
penguins = sns.load_dataset('penguins')
````

````python
penguins.head()
````

````python
# H registros missing?
penguins.isna().sum()
````

````python
# Onde esto os missing?
penguins[penguins.body_mass_g.isna()]
````

````python
# Eliminando os missing
penguins.dropna(subset=['body_mass_g'], inplace=True)
````

````python
# selecionando apenas variveis numricas
penguins_quant = penguins.select_dtypes(include="number")
penguins_quant.head()
````

## Anlise descritiva
Antes de rodar o algoritmo, vamos avaliar uma matriz de grficos de disperso (*scatterplot matrix*) para verificar se h algum padro evidente nos dados.

````python
sns.pairplot(penguins, hue='species')
````

## Rodando o algoritmo
De uma forma bem similar a outros algoritmos de *machine learning*, vamos definir um objeto com a classe KMeans, em seguida vamos aplicar o mtodo fit, indicando a base de dados.

````python
kmeans = KMeans(n_clusters=3, max_iter=600, algorithm = 'lloyd', n_init='auto')
kmeans.fit(penguins_quant)
````

Podemos identificar os grupos com o atribuito ```labels_```. Vamos construir um grfico de disperso para ver como esses grupos ficaram.

````python
penguins['grupo'] = kmeans.labels_.astype('str')
sns.pairplot(penguins, hue='grupo')
````

Os grupos ficaram um tanto esuquisitos... os grficos de baixo esto com um aspecto de "bandeira". O que aconteceu  que o algoritmo utilizou quase que somente a varivel ```body_mass```. Por que? 

Bem, repare na escala dessa varivel, ela est em gramas e fornece valores muito maiores que as demais variveis. Faa o exerccio de calcular a distcia entre um par de pinguins, e voc vai confirmar que a contribuio da varivel ```body_mass```  muito maior que as demais variveis.

## Distncia

Para fazer o agrupamento, o algoritmo usa uma medida de *dissimilaridade* ou simplesmente distncia (o quanto duas observaes so diferentes) para agrupar os pinguins. A medida mais comum, a que usaremos,  a distncia euclidiana calculada nas variveis quantitativas entre dois indivduos (```a``` e ```b```) pode ser escrita como:

$$ d = \sqrt{\sum_{i=1}^{p} (\delta(x_i))^2} $$

com:
- $x_i$ sendo a i-sima varivel
- i indexando as variveis, valendo 1, 2, ... p
- p  o nmero de variveis
- $\delta(x_i) = x_{i, a} - x_{i, b}$  a diferena da varivel $x_i$ entre os indivduos a e b.

## Variveis em escalas diferentes
Variveis com unidades de medida diferentes frequentemente fornecem escalas diferentes, como no caso dos pinguins. Quando calculamos uma distncia considerando estas variveis to diferentes, estamos realizando uma "ponderao arbitrria".
- **Ponderao** porque estamos dando pesos ou importncias diferentes para cada varivel
- **Arbitrria** porque podemos medir a massa corporal em kilos ou toneladas, sem alterar a informao que a varivel traz, mas alterando **arbitrariamete** a importncia que ela teria no algoritmo.

## Como resolver o problema de escala?

Ok, como resolvemos isso? 

Uma forma muito popular de abordar este problema  realizar uma alterao de escala padronizada (um *scaling* em ingls) nas variveis quantitativas. Para qual escala podemos transformar? Se dividirmos cada varivel pelo seu desvio padro, a unidade de medida da varivel fica 'adimensional' - ou podemos entender que uma unidade equivale a 1 desvio padro (vale refletir a respeito). Costuma-se tambm subtrair a prpria mdia da varivel, de modo que ela fique com mdia=0 e desvio padro = 1.

Essa padronizao pode ser escrita assim:

$$ x_{pad} = \frac{x - \bar{x}}{desv\_pad(x)} $$

A funo ```StandardScaler``` faz isso de forma bem prtica.

````python
padronizador = StandardScaler()
penguin_padronizado = padronizador.fit_transform(penguins_quant)
````

````python
sns.pairplot(pd.DataFrame(penguin_padronizado))
````

````python
kmeans.fit(penguin_padronizado)
````

Vamos ver agora como os grupos ficaram:

````python
penguins['grupo'] = kmeans.labels_
sns.pairplot(penguins, hue='grupo')
````

Parece fazer bem mais sentido esse agrupamento. Ele parece considerar igualmente todas as variveis. Vamos checar abaixo como fica a mesma anlise se realarmos no ```hue``` as espcies de pinguim.

````python
penguins['grupo'] = kmeans.labels_
sns.pairplot(penguins, hue='species')
````

````python
pd.crosstab(penguins['species'], penguins['grupo'])
````

Os grupos tm uma concordncia boa com as espcies! Mas no se esquea: no estamos tentando 'acertar' as espcies, mas sim encontrar padres entre as variveis biomtricas dos pinguins e dividi-los em grupos de pinguins *semelhantes dentro do grupo e diferentes entre grupos*. A concordncia  uma curiosa e feliz coincidncia.
