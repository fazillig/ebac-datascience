````python
metadados = pd.DataFrame([df[i].dtype for i in df.columns], columns=['tipo'], index=df.columns)
metadados.index.name = "Coluna"
metadados['qtd_categorias'] = metadados.index.map(lambda i: df[i].nunique())

metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df.head()
````

````python
df_dummies = pd.get_dummies(df, columns=['sexo', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'])
df_dummies.posse_de_veiculo = df_dummies.posse_de_veiculo.map({'Y': 1, 'N': 0})
df_dummies.posse_de_imovel = df_dummies.posse_de_imovel.map({'Y': 1, 'N': 0})
df_dummies.posse_de_veiculo = df_dummies.posse_de_veiculo.astype(int)
df_dummies.posse_de_imovel = df_dummies.posse_de_imovel.astype(int)
df_dummies.mau = df_dummies.mau.astype(int)
df_dummies.info()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
round(pd.crosstab(df_dummies.possui_email, df_dummies.mau, normalize=True)*100,2)
````

````python
round(pd.crosstab(df_dummies.posse_de_veiculo, df_dummies.mau, normalize=True)*100,2)
````

````python
df_dummies[['posse_de_veiculo', 'possui_email', 'mau']].corr()
````

**Observação:**
- Existem mais pessoas (0,86%) que possuem veículo e são mau pagadoras do que pessoas que são mas pagadoras e possuem email (0,18%).
- Existe uma correlação fraca dessas variáveis com a variável mau, porém, entre as duas a variáveis, a posse_de_veiculo tem uma correlação negativa mais forte.

Por tanto a variável posse_de_veiculo aparenta ser melhor para prever
