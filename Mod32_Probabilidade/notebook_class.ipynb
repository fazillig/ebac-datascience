## Gerando nmeros pseudoaleatrios

Gerar nmeros aleatrios tem uma infinidade de utilidades. Os nmeros aleatrios mais comuns so os que possuem distribuio uniforme entre 0 e 1. Ou seja, so nmeros contnuos, gerados pseudoaleatoriamente, com igual probabilidade de cair em cada segmento do intervalo que tenha mesmo tamanho.

O Numpy tem um pacote grande de nmeros pseudoaleatrios, que geram valores uniformes.

````python
np.random.rand(10)
````

````python
fig, ax = plt.subplots(figsize=(6,4))

ax.hist(np.random.rand(100000), density=True)

ax.yaxis.set_major_formatter(mtick.PercentFormatter(10))
````

## Distribuio conjunta

As probabilidades de duas variveis aleatrias podem ser 

````python
df = pd.DataFrame({'Dado1':np.random.randint(1, 7, 10000),'Dado2':np.random.randint(1, 7, 10000)})
df['Soma'] = df['Dado1']+ df['Dado2']
df.head()
````

## Independncia

Em probabilidade, duas variveis so independentes se $P(X1=x1|X2=x2) = P(X1=x1)$. Ou seja, $X2$ assumir um determinado valor no d informao alguma sobre a probabilidade de $X1$ assumir algum valor. OBS: novamente estamos denotando variveis aleatrias com letras maisculas, e valores especficos com letras minsculas.

Vamos ilustrar com os resultados dos dois dados gerados acima. Por exemplo, $P(Dado2|Dado1=1)=P(Dado2)$. Ou seja, a probabilidade de o dado 2 assumir algum resultado  independente do resultado do dado 1.

````python
df[df['Dado1'] == 5].loc[:,'Dado2'].plot.hist(bins=6)
````

## Distribuio Condicional

Para variveis que possuem alguma dependncia, por exemplo, $X1$, o resultado do primeiro dado e $Y$, a soma dos dois dado, podemos nos interessar pela distribuio condicional. Isto , se o primeiro dado resultar em um valor especfico, qual a funo de probabilidade da soma?

````python
df[df['Dado1'] == 6].loc[:,'Soma'].plot.hist(bins=6)
````

## Funo densidade de probabilidade

A funo densidade de probabilidade, $f(x)$, descreve a 'lei' com que uma varivel contnua  gerada. Embora a probabilidade de uma varivel contnua assumir qualquer valor pontual seja zero ( $P(X=x)=0$ para qualquer x), a probabilidade de X ocorrer em um intervalo $[a, b]$, por exemplo, pode ser descrita pela rea abaixo da curva nesse intervalo, ou seja, a integral $\int_{a}^{b} f(x) \,dx$.

Vamos fazer o grfico de alguns exemplos para ilustrar:

````python
fig, ax = plt.subplots(figsize=(10,3))

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
ax.plot(x, y);
````

````python
fig, ax = plt.subplots(figsize=(10,2))

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
ax.plot(x, y)

std_lim = 1 # 95% CI
low = 0
high = mu+std_lim*sigma

plt.fill_between(x, y, where=(low < x) & (x < high), alpha=.5)
# plt.text(low, 0, low, ha='center')
# plt.text(high, 0, high, ha='center')

prob_area = norm.cdf(high, mu, sigma) - norm.cdf(low, mu, sigma)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))
````

## Funes Distribuio Acumulada

A funo distribuio acumulada, como o nome diz,  a verso acumulada da funo densidade de probabilidade (ou da funo de probabilidade para variveis discretas). Ela d a seguinte probabilidade: $F(x) = P(X<x)$ - nessa expresso, denotamos varivel aleatria com letra maiscula ($X$) e um valor constante arbitrrio com letra minscula ($x$).

Vamos ver isso graficamente:

````python
fig, ax = plt.subplots(2, 1, figsize=(10,4))

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
cdf = stats.norm.cdf(x, mu, sigma)
ax[1].plot(x, cdf)

low = -np.inf
high = 3


ax[1].yaxis.set_major_formatter(mtick.PercentFormatter(1))
ax[1].vlines(x=[high], 
             ymin=0, ymax=stats.norm.cdf(high, mu, sigma), 
             colors='r', linestyles='dotted')
ax[1].hlines(y=stats.norm.cdf(high, mu, sigma), 
             xmin=mu - 3*sigma, xmax=high, 
             colors='r', linestyles='dotted')

prob_area = norm.cdf(high, mu, sigma) - norm.cdf(low, mu, sigma)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))

y = stats.norm.pdf(x, mu, sigma)
ax[0].plot(x, y)
ax[0].fill_between(x, y, where=(low < x) & (x < high), alpha=.5)
# plt.text(low, 0, low, ha='center')
# plt.text(high, 0, high, ha='center')
````

 muito comum utilizar, em linguagens de programao, a funo distribuio acumulada ($F(x)$) para se obter a probabilidade de uma varivel aleatria estar em determinado intervalo. Para se obter a probabilidade da varivel aleatria $X$ estar no intervalo $[a, b]$ fazemos $F(b)-F(a)$.

Por exemplo, vamos calcular a probabilidade de uma varivel aleatria com distribuio Normal Padro estar no intervalo $[-1,0]$ e mostrar isso no grfico:

````python
fig, ax = plt.subplots(figsize=(10,2))

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.cdf(x, mu, sigma)
ax.plot(x, y)

std_lim = 0 # 95% CI
low = -.05
high = .05

plt.fill_between(x, y, where=(low < x) & (x < high), alpha=.5)
# plt.text(low, 0, low, ha='center')
# plt.text(high, 0, high, ha='center')

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1))

prob_area = norm.cdf(high, mu, sigma) - norm.cdf(low, mu, sigma)
print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.2%}'.replace('.',','))
````

A funo distribuio acumulada  muito utilizada nas linguagens de programao. Quando temos uma varivel aleatria com uma distribuio conhecida (e implementada no numpy, por exemplo), podemos obter a probabilidade de uma realizao dessa varivel pertencer a qualquer intervalo.

## Distribuio *t de studdent*

````python
fig, ax = plt.subplots(figsize=(10,4))

n = 5
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
ax.plot(x, y, label='Normal')

t = stats.t.pdf(x, n-1)
ax.plot(x, t, label=f't({n}-1)');


ax.legend()
#
````

````python
fig, ax = plt.subplots(figsize=(10,4))

mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)
ax.plot(x, y, label='Normal')

for gl in ([1, 2, 10]):
    t = stats.t.pdf(x, gl)
    ax.plot(x, t, label=f't({gl})');


ax.legend()
````

````python
# Parmetros da distribuio
mu = 0
variance = 1
gl = 1

#varincia
sigma = math.sqrt(variance)

# Calcular probabilidade para o intervalo low e high
low = -np.inf
high = 1

# X distribuido uniformemente
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
cdf = stats.t.cdf(x, gl)*sigma+ mu

# Figura
fig, ax = plt.subplots(2, 1, figsize=(10,4))

# Plot da densidade
y = stats.t.pdf(x, gl)*sigma+ mu
ax[0].plot(x, y)
# Realar a rea entre LOW e HIGH
ax[0].fill_between(x, y, where=(low < x) & (x < high), alpha=.5)

# Plot da distribuio acumulada
ax[1].plot(x, cdf)

# Formatar y da CDF com percentual
ax[1].yaxis.set_major_formatter(mtick.PercentFormatter(1))
# Linhas de referncia 
ax[1].vlines(x=[high], 
             ymin=0, ymax=stats.t.cdf(high, gl)*sigma+ mu, 
             colors='r', linestyles='dotted')
ax[1].hlines(y=stats.t.cdf(high, gl)*sigma+mu, 
             xmin=mu - 3*sigma, xmax=high, 
             colors='r', linestyles='dotted')
ax[1].set_ylim([0, 1])

# Calcular probabilidade entre LOW e HIGH
prob_area = stats.t.cdf(high, gl)*sigma - stats.t.cdf(low, gl)*sigma
plt.show()

print(f'A probabilidade para valores entre {low:.2f} e {high:.2f}  de {prob_area:.1%}'
      .replace('.',','));
````

## Explorando funes de probabilidade

Diversas funes de distribuio esto implementadas no SciPy em classes. O que significa 'estarem implementadas em classes'? Significa que, como em todos os objetos do Python, objetos dessas classes vo possuir atributos e mtodos que sero teis para as nossas atividades.

Vamos entender. Na clula abaixo, por exemplo, estamos chamando direto do SciPy a classe ```t```, e dela o mtodo ```cdf```, com um parmetro ```x```, que recebe lista, e um parmetro ```gl``` que recebe um nmero. O mtodo vai retornar um valor para cada elemento do ```x```.

````python
from scipy import stats

gl=1
stats.t.cdf([-1, 0, 1, 2, 7, 100], gl)
````

Na construo abaixo, estamos criando um objeto ```t1```, na classe ```stats.t```, com o parmetro de graus de liberdade "congelado" com o valor 1.

Em seguida, estamos utilizando o mtodo ```cdf``` para calcular a distribuio acumulada (*cumulative distribution function*) para cada valor da lista atribuda ao parmetro ```x``` do mtodo. 

````python
t1 = stats.t(1)
t1.cdf([-1, 0, 1, 2])
````

### Inversa da t
O SciPy possui ainda uma variedade de mtodos. Por exemplo, possui o mtodo ```ppf``` - de *percent point function*, que retorna o inverso da ```cdf```: inserindo uma probabilidade (ou um percentil), o mtodo retorna o valor de ```x``` correspondente a esta probabilidade segundo a ```cdf```.

````python
t1.ppf([.05, .25, .5, .75, .95])
````
